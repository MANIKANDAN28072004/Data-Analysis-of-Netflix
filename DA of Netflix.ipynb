```python
import numpy as np
import pandas as pd
import plotly.express as px
from textblob import TextBlob

df=pd.read_csv('netflix_titles.csv')
```

```python
x=df.groupby(['rating']).size().reset_index(name='counts')
print(x)
```

```python
piechart=px.pie(x,values='counts',names='rating',title='Distribution of Ratings in Netflix')
piechart.show()
```

```python
df['director']=df['director'].fillna('Director not specified')
df.head()
```

```python
director_list=pd.DataFrame()
print(director_list)
```

```python
director_list=df['director'].str.split(',',expand=True).stack()
print(director_list)
```

```python
director_list.columns=['Director']
print(director_list)

```

```python
directors=director_list.groupby(['Director']).size().reset_index(name='Tot Count')
print(directors)
```

```python
directors=directors[directors.Director != 'Director not specified']
print(directors)
```

```python
directors=directors.sort_values(by='Tot Count',ascending=False)
print(directors)
```

```python
top5director = directors.head()
print(top5director)
```

```python
top5director=top5director.sort_values(by=['Tot Count'])
bargraph=px.bar(top5director,x='Tot Count',y='Director',title='Top 5 Directors in Netflix')
bargraph.show()
```

```python
df['cast']=df['cast'].fillna('cast not specified')
cast_df=pd.DataFrame()
cast_df=df['cast'].str.split(',',expand=True).stack()
cast_df=cast_df.to_frame()
cast_df.columns=['Actor']
actors=cast_df.groupby(['Actor']).size().reset_index(name='Tot Count')
actors=actors[actors.Actor != 'cast not specified']
actors=actors.sort_values(by=['Tot Count'],ascending=False)
top5actors=actors.head()
top5actors=top5actors.sort_values(by=['Tot Count'])
bargraph2=px.bar(top5actors,x='Tot Count',y='Actor',title='Top 5 Actors in the Netflix')
bargraph2.show()
```

```python
df1=df[['type','release_year']]
df1=df1.rename(columns={"release_year":"Release Year","type":"Type"})
df2=df1.groupby(['Release Year','Type']).size().reset_index(name='Tot Count')
```

```python
print(df2)
```

```python
df2=df2[df2['Release Year']>=2000]
graph=px.line(df2,x='Release Year',y='Tot Count',color='Type',title='Trend of the Content Produced in Netflix Every Year')
graph.show()
```

```python
df['country']=df['country'].fillna('country not specified')
county_df=pd.DataFrame()
county_df=df['country'].str.split(',',expand=True).stack()
county_df=county_df.to_frame()
county_df.columns=['Country']
nadu=county_df.groupby(['Country']).size().reset_index(name='Tot Count')
nadu=nadu[nadu.Country != 'country not specified']
nadu=nadu.sort_values(by=['Tot Count'],ascending=False)
top5countries=nadu.head()
top5countries=top5countries.sort_values(by=['Tot Count'])
bargraph2=px.bar(top5countries,x='Tot Count',y='Country',title='Top 5 Countries on the Netflix')
bargraph2.show()
```

```python
df3 = df[['release_year', 'description']]
df3 = df3.rename(columns = {'release_year':'Release Year', 'description':'Description'})
for index, row in df3.iterrows():
  d=row['Description']
  testimonial = TextBlob(d)
  p = testimonial.sentiment.polarity
  if p==0:
    sent = 'Neutral'
  elif p>0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df3.loc[[index, 2], 'Sentiment']=sent

df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name = 'Tot Count')

df3 = df3[df3['Release Year']>2005]
barGraph = px.bar(df3, x="Release Year", y="Tot Count", color = "Sentiment", title = "Sentiment Analysis of Content on Netflix")
barGraph.show()
```
